```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from scipy import stats
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.utils import to_categorical
from keras.layers import LSTM, Dense, Input
```

 Load the datasets

```python
red_wine = pd.read_csv('/content/winequality-red.csv', sep=';')
white_wine = pd.read_csv('/content/winequality-white.csv', sep=';')
```

Dataset Description

```python
print("Red Wine Dataset:")
print(red_wine.head())
print("\nWhite Wine Dataset:")
print(white_wine.head())

print("\nRed Wine Dataset Info:")
print(red_wine.info())
print("\nWhite Wine Dataset Info:")
print(white_wine.info())
```

Combine Datasets

```python
red_wine['wine_type'] = 'red'
white_wine['wine_type'] = 'white'
wine_data = pd.concat([red_wine, white_wine], ignore_index=True)
```

Check for Missing Values

```python
print("\nMissing Values in Each Column:")
print(wine_data.isnull().sum())
```

Handle Outliers

```python
def remove_outliers_zscore(df, threshold=3):
    z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
    return df[(z_scores < threshold).all(axis=1)]

wine_data_clean = remove_outliers_zscore(wine_data)

print(f"\nOriginal dataset size: {wine_data.shape}")
print(f"Dataset size after outlier removal: {wine_data_clean.shape}")
```

Exploratory Data Analysis

```python
wine_data_clean.hist(bins=20, figsize=(15,10))
plt.suptitle('Histograms of Wine Features')
plt.show()
```

 Scatter Plots

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='alcohol', y='quality', data=wine_data_clean, hue='wine_type')
plt.title('Alcohol vs Quality')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='volatile acidity', y='quality', data=wine_data_clean, hue='wine_type')
plt.title('Volatile Acidity vs Quality')
plt.show()
```

Correlation Matrix

```python
# Exclude the 'wine_type' column from the correlation matrix since it's categorical
plt.figure(figsize=(12,10))

# Select only numeric columns for the correlation matrix
corr_matrix = wine_data_clean.select_dtypes(include=[np.number]).corr()

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

 Machine Learning Model (Random Forest)

```python
X = wine_data_clean.drop(['quality', 'wine_type'], axis=1)
y = wine_data_clean['quality']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Initialize and train Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# Evaluate Random Forest model
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Random Forest Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_rf) * 100))
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
y_pred_mlp = mlp.predict(X_test)
```

```python
# Evaluate MLP model
print("MLP Classification Report:\n", classification_report(y_test, y_pred_mlp))
print("MLP Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_mlp) * 100))
```

```python
# --- Neural Network Model (LSTM) ---
# Reshape data for LSTM
X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Convert target labels to categorical
num_classes = np.max(y_train) + 1  # Ensure num_classes reflects the maximum label + 1
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)
```

```python
# Build LSTM model
model = Sequential()
# Use Input layer to specify input shape
model.add(Input(shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
model.add(LSTM(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Compile and train the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_reshaped, y_train_cat, epochs=50, batch_size=64, validation_data=(X_test_reshaped, y_test_cat), verbose=2)
```

```python
# Evaluate LSTM model
loss, accuracy = model.evaluate(X_test_reshaped, y_test_cat, verbose=0)
print("LSTM Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
# --- Model Comparison ---
rf_accuracy = accuracy_score(y_test, y_pred_rf) * 100
mlp_accuracy = accuracy_score(y_test, y_pred_mlp) * 100
lstm_accuracy = accuracy * 100

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Model': ['Random Forest', 'MLP', 'LSTM'],
    'Accuracy (%)': [rf_accuracy, mlp_accuracy, lstm_accuracy]
})
```

```python
print("\nModel Accuracy Comparison:")
print(comparison_df)
```

```python
# Plot comparison
plt.figure(figsize=(10,6))
sns.barplot(x='Accuracy (%)', y='Model', data=comparison_df, palette='viridis')
plt.title('Model Comparison Based on Accuracy')
plt.xlabel('Accuracy (%)')
plt.ylabel('Model')
plt.xlim(0, 100)
plt.show()
```

```python
# --- Random Forest Hyperparameter Tuning ---
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

print("Best Parameters for Random Forest:", grid_search_rf.best_params_)
best_rf = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf.predict(X_test)

print("\nBest Random Forest Classification Report:\n", classification_report(y_test, y_pred_best_rf))
print("Best Random Forest Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_best_rf) * 100))
```

```python
# --- MLP Hyperparameter Tuning (Reduced) ---
param_grid_mlp = {
    'hidden_layer_sizes': [(50, 50)],  # Start with one configuration
    'activation': ['relu'],
    'solver': ['adam'],
    'alpha': [0.0001],
    'learning_rate': ['constant'],
}

# Initialize GridSearchCV with fewer parameters
grid_search_mlp = GridSearchCV(estimator=MLPClassifier(max_iter=500, random_state=42),
                               param_grid=param_grid_mlp,
                               cv=3,  # Fewer folds
                               n_jobs=-1,
                               verbose=2,
                               scoring='accuracy')

# Fit the model
grid_search_mlp.fit(X_train, y_train)

print("Best Parameters for MLP:", grid_search_mlp.best_params_)
best_mlp = grid_search_mlp.best_estimator_
y_pred_best_mlp = best_mlp.predict(X_test)

print("\nBest MLP Classification Report:\n", classification_report(y_test, y_pred_best_mlp))
print("Best MLP Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_best_mlp) * 100))
```

```python
# --- Final Model Comparison ---
comparison_df_final = pd.DataFrame({
    'Model': ['Random Forest', 'Best Random Forest', 'MLP', 'Best MLP', 'LSTM'],
    'Accuracy (%)': [
        rf_accuracy, accuracy_score(y_test, y_pred_best_rf) * 100,
        mlp_accuracy, accuracy_score(y_test, y_pred_best_mlp) * 100,
        lstm_accuracy
    ]
})

print("\nFinal Model Accuracy Comparison:")
print(comparison_df_final)
```

```python
# Plot Final comparison
plt.figure(figsize=(10,6))
sns.barplot(x='Accuracy (%)', y='Model', data=comparison_df_final, palette='viridis')
plt.title('Final Model Comparison Based on Accuracy')
plt.xlabel('Accuracy (%)')
plt.ylabel('Model')
plt.xlim(0, 100)
plt.show()
```
